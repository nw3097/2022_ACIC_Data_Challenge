```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn import metrics, preprocessing
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from final_submission_helper import *
pd.set_option('display.max_columns', None)
```

#### run loop

```python
%%time
# for actual run, set max_range = 3401
max_range = 3
submission_df_dml_with_weighting = pd.DataFrame()
for data_set_number in range(1, max_range):
    padded_seq_no = f'{data_set_number:04}'
    df1, df2, df3, df4, df = load_data(data_set_number)
    df3_PS = build_ps_df(df3, df4, df, quantile_granularity=5)
    gbm = do_lgbm_ps_model(df3_PS, ps_features, cat_features, 'Z')
    df3_PS = add_ps_weights(df3_PS, gbm, ps_features)
    full_data = setup_estimation_data(df, df3_PS)
    res_df = do_dml_full_pipeline_with_weighting(full_data)
    res_df['dataset.num'] = pd.Series(padded_seq_no)
    res_df = res_df[['dataset.num','variable','level','year','satt','lower90','upper90']]
    submission_df_dml_with_weighting = submission_df_dml_with_weighting.append(res_df)
```
